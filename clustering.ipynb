```python
#import libraries
import numpy as np
from sklearn.cluster import KMeans
```

```python
#image names
image_names = [

    "dog_1.png","dog_2.png","dog_3.png",
    "dog_4.png","dog_5.png","dog_6.png",

    "cat_1.png","cat_2.png","cat_3.png",
    "cat_4.png","cat_5.png","cat_6.png",

    "car_1.png","car_2.png","car_3.png",
    "car_4.png","car_5.png","car_6.png",

    "mountain_1.png","mountain_2.png","mountain_3.png",
    "mountain_4.png","mountain_5.png","mountain_6.png",

    "unknown1.png",
    "unknown2.png",
    "unknown3.png",
    "unknown4.png",
    "unknown5.png",
    "unknown6.png"

]
```

```python
# ----------------------------------------------------------
# Sample Image Features
#
# Suppose every image is converted into two numerical
# features:
#
# Feature 1 = Color
# Feature 2 = Texture
#
# These values are made up ONLY for demonstration.
# ----------------------------------------------------------
# Our Algorithm will not know if it's dog, cat, cars, Mountain, Unkown Image
# Unlabelled , 

image_features = np.array([

# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],

# Unknown Images
[250,50],     # Flower
[180,20],     # Airplane
[170,100],    # Bicycle
[90,150],     # City
[230,90],     # Burger
[180,200]     # Beach

])
```

```python
# Create KMeans Model
# ----------------------------------------------------------

kmeans = KMeans(
    n_clusters=4,
    random_state=42
)

# Train the model

kmeans.fit(image_features)

# Cluster assignments

clusters = kmeans.labels_
```

```python
# Display Results
# ----------------------------------------------------------

print("IMAGE CLUSTERING RESULTS")


for image, cluster in zip(image_names, clusters):

    print(f"{image:18} ---> Cluster {cluster}")

# ----------------------------------------------------------
# Cluster Centers
# ----------------------------------------------------------

print("\nCluster Centers")
print("-"*40)

print(kmeans.cluster_centers_)
```
